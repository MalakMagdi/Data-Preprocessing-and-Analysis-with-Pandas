`['Country', 'Energy Supply', 'Energy Supply per Capita', '% Renewable']`

Convert `Energy Supply` to gigajoules (there are 1,000,000 gigajoules in a petajoule). For all countries which have missing data (e.g. data with "...") make sure this is reflected as `np.NaN` values.

Rename the following list of countries (for use in later questions):

```"Republic of Korea": "South Korea",

````python
import pandas as pd
import numpy as np
````

````python
#%pip install xlrd
````

````python
Energy = pd.read_excel('Energy Indicators.xls')
````

````python
#Remove Head and Footer 
Energy = pd.read_excel('Energy Indicators.xls', header=17, skipfooter=38)
````

````python
#check that footer is removed 
Energy.tail(10)
````

````python
# Remove the first column 
Energy.drop('Unnamed: 0', inplace=True, axis=1)
````

````python
#Removing Second Column 
Energy.drop('Unnamed: 1', inplace=True, axis=1)
````

````python
#Renaming The Columns 
Mod_Energy = Energy.rename(columns={'Unnamed: 2': 'Country', 'Petajoules': 'Energy Supply', 'Gigajoules': 'Energy Supply per Capita','%': '% Renewable'})
````

````python
#Multiply Energy Supply to be in Gigajoules
Mod_Energy['Energy Supply'] = Mod_Energy['Energy Supply'].apply(lambda x: x * 1000000)
````

````python
#Change missing values to np.NaN
Mod_Energy.replace(to_replace = '...', value =np.NaN , inplace=True)
````

````python
# To confirm the precense of NaN 
Mod_Energy.isnull().sum()
````

````python
# Removing Parenthesis and Numbers from the data
Mod_Energy['Country'] = Mod_Energy['Country'].str.replace('\d+', '' , regex=True)
Mod_Energy['Country'] = Mod_Energy['Country'].str.replace(r" \(.*\)","" , regex=True)
#Testing 
# Mod_Energy.style
````

````python
#Replace Values of some countries with new values
Mod_Energy.replace({'Country' : {'Republic of Korea': 'South Korea','United States of America': 'United States','United Kingdom of Great Britain and Northern Ireland': 'United Kingdom',
'China, Hong Kong Special Administrative Region': 'Hong Kong' }}, inplace=True)
#Testing
(Mod_Energy['Country'].eq('Hong Kong')).any()
````

````python
#Checking the data after cleaning it 
Energy=Mod_Energy
display(Energy)
````

### Step 3
Join the three datasets: GDP, Energy, and ScimEn into a new dataset (using the intersection of country names). Use only the last 10 years (2006-2015) of GDP data and only the top 15 countries by Scimagojr 'Rank' (Rank 1 through 15). 

The index of this DataFrame should be the name of the country, and the columns should be ['Rank', 'Documents', 'Citable documents', 'Citations', 'Self-citations',

       '2009', '2010', '2011', '2012', '2013', '2014', '2015'].

*This step should yeild a DataFrame with 20 columns and 15 entries.*

````python
#Joining the data frames 
Energy_GDP_Merged = pd.merge(Energy, 
                             GDP [['2006', '2007', '2008', '2009', '2010', '2011', '2012', '2013', '2014', '2015','Country Name']], 
                             how='left', 
                             left_on='Country', 
                             right_on='Country Name')

Final=pd.merge(ScimEn,
            Energy_GDP_Merged,
              how='inner',
              on='Country')

Total_Row_Count=len(Final.index)

#Get only 15 row 
Final.drop(Final.index[15: ], inplace=True)

Final_Row_Count= len(Final.index)

#Set Country as the index
Final.set_index('Country',inplace=True)

#Remove Repeated Columns
Final.drop('Country Name', inplace=True, axis=1)

#Test
Final.head()
Final.columns
````
